```
yaml_config="""
input_features:
    -
        name: Original_Title
        type: text
        level: word
        encoder: t5
        reduce_output: null
    -
        name: Keyword
        type: text
        level: word
        tied_weights: Original_Title
        encoder: t5
        reduce_output: null
output_features:
    -
        name: Optimized_Title
        type: sequence
        level: word
        decoder: generator
"""
```

```
with open("config.yaml", "w") as f:
  f.write(yaml_config)
```

```
!ludwig train --dataset hootsuite_titles.csv --config ludwig_t5_config.yaml.1
```
